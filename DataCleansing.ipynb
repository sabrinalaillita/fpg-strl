```python
#pip install pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import dataset
df = pd.read_csv("Sales.csv")
print(df.head(5))
```

```python
# Change 'Tanggal' data type to Datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
df.info()
```

```python
# Deleteing rows with 'status' = 'belum lunas'
df_clean = df.copy()
print(df_clean[df_clean['Status'].str[:1]=='B'].count())
df_clean = df_clean[df_clean['Status'].str[:1]!='B']
```

```python
# Deleteing rows with 'HeaderReturFaktur' > 0
print(df_clean[df_clean['HeaderReturFaktur']>0].count())
df_clean = df_clean[df_clean['HeaderReturFaktur']==0]
```

Data Cleansing

```python
# Check for missing value
df_clean.isnull().sum()
```

```python
# Check for duplicated data
print("Jumlah duplikasi: ", df_clean.duplicated().sum())
```

```python
# Resolving data inconsistencies

df_clean['Outlet'] = df_clean['Outlet'].str.title()
df_clean['NomorFaktur'] = df_clean['NomorFaktur'].str.upper()
df_clean['Status'] = df_clean['Status'].str.title()
df_clean['SKU_Produk'] = df_clean['SKU_Produk'].str.upper()
df_clean['NamaProduk'] = df_clean['NamaProduk'].str.title()
```

Exploratory Data Analysis (EDA)

```python
# Find the 5 top-selling products
# Group by product and sum the quantities
sales_by_product = df_clean.groupby('NamaProduk')['Qty'].sum()

# Sort the products by total sales and select the top 5
top_5_products = sales_by_product.sort_values(ascending=False).head(5)
print(top_5_products)
```

```python
# Find the worst-selling products
# Group by product and sum the quantities
sales_by_product = df_clean.groupby('NamaProduk')['Qty'].sum()

# Sort the products by total sales and select the top 5
worst_selling_products = sales_by_product[sales_by_product <= 1]
print(worst_selling_products)
```

```python
# Find the top-selling Outlet
# Group by product and sum the quantities
sales_by_outlet = df_clean.groupby('Outlet')['Qty'].sum()

# Sort the outlets by total sales
sorted_sales_by_outlet = sales_by_outlet.sort_values(ascending=False)

# Best-selling outlet
best_selling_outlet = sorted_sales_by_outlet.idxmax()
print("Best-selling Outlet: " + best_selling_outlet)

# Worst-selling outlet
worst_selling_outlet = sorted_sales_by_outlet.idxmin()
print("Worst-selling Outlet: " + worst_selling_outlet)
```

```python
# Create a bar chart
def set_custom_palette(series, max_color = 'royalblue', min_color = "lightcoral", other_color = 'lightgrey'):
    max_val = series.max()
    min_val = series.min()
    pal = []
    
    for item in series:
        if item == max_val:
            pal.append(max_color)
        elif item == min_val:
            pal.append(min_color)
        else:
            pal.append(other_color)
    return pal

plt.figure(figsize=(12,9))
ax = sns.barplot(x='Qty', y='Outlet', hue="Outlet", legend=False,
            data=sales_by_outlet, 
            palette=set_custom_palette(sales_by_outlet['Qty']))
for index, value in enumerate(sales_by_outlet['Qty']):
    plt.text(value + 0.5, index, f'{value}', va='center', ha='left', fontsize=10, color='black')
ax.set_title('Total Sales by Outlet')
ax.set(xlabel='Total Sales', ylabel= 'Outlet');
```

Convert Data into a format suitable for Market Basket Analysis

```python
transactions = df_clean.groupby('NomorFaktur')['NamaProduk'].apply(list)

# Converting to a list of transactions
transactions_list = transactions.tolist()

# Display the first few transactions
print(transactions_list[:5])
```

Perform Market Basket Analysis Using mlxtend

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth

# Step 1: Transform the transaction list into the right format
te = TransactionEncoder()
te_ary = te.fit(transactions_list).transform(transactions_list)
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)

# Step 2: Apply the FP-Growth algorithm
frequent_itemsets = fpgrowth(df_transactions, min_support=0.01, use_colnames=True)

# Display the frequent itemsets
print(frequent_itemsets)

# TransactionEncoder: Converts the list of transactions into a one-hot encoded DataFrame, where each product is a column, and each row corresponds to a transaction.
# fpgrowth: This function applies the FP-Growth algorithm to find frequent itemsets that meet the minimum support threshold.
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# transactions_list: A list of transactions where each transaction is a list of products bought together.
# fpgrowth: Extracts frequent itemsets.
# association_rules: Generates rules with metrics like support, confidence, and lift.
```

Data Visualization of The Result

```python
# Create a new column that converts the frozenset to a readable string
frequent_itemsets['itemsets_str'] = frequent_itemsets['itemsets'].apply(lambda x: ', '.join(list(x)))

# Sort the itemsets by support
top_itemsets = frequent_itemsets.sort_values(by='support', ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='support', y='itemsets_str', hue='itemsets_str', legend=False, data=top_itemsets, palette='RdYlBu')
for index, value in enumerate(top_itemsets['support']):
    plt.text(value, index, round(value, 3), va='center', ha='left', fontsize=10, color='black')
plt.title('Top 10 Frequent Itemsets by Support', fontsize=14)
plt.xlabel('Support', fontsize=12)
plt.ylabel('Itemsets', fontsize=12)
plt.show()
```

```python
# Create a matrix for the heatmap
# The index will be the 'antecedents' and the columns will be the 'consequents'
# The values will be the 'lift'
heatmap_data = rules.pivot(index='antecedents', columns='consequents', values='lift')

# Convert frozenset to string for better readability in the heatmap
heatmap_data.index = heatmap_data.index.map(lambda x: ', '.join(list(x)))
heatmap_data.columns = heatmap_data.columns.map(lambda x: ', '.join(list(x)))

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", linewidths=.5)
plt.title('Heatmap of Itemset Correlations (Lift)')
plt.show()
```

```python
import networkx as nx

# Create a directed graph from the rules
G = nx.DiGraph()

# Add nodes for each unique item in the rules
for _, rule in rules.iterrows():
    for item in rule['antecedents']:
        G.add_node(item)
    for item in rule['consequents']:
        G.add_node(item)
        
    # Add edges with lift as the weight
    G.add_edges_from([(antecedent, consequent, {'lift': rule['lift'], 
                                                'confidence': rule['confidence'], 
                                                'support': rule['support']}) 
                      for antecedent in rule['antecedents'] 
                      for consequent in rule['consequents']])

# Draw the graph
plt.figure(figsize=(15, 15))
pos = nx.spring_layout(G, k=0.5, iterations=50)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, 
        font_size=10, font_weight='bold', edge_color='gray')

# Draw edges with varying thickness based on lift value
edges = G.edges(data=True)
nx.draw_networkx_edges(G, pos, edgelist=edges, arrowstyle='-|>', arrowsize=20,
                       edge_color=[e[2]['lift'] for e in edges], 
                       edge_cmap=plt.cm.Blues, width=2)

# Add edge labels (optional: showing lift)
edge_labels = {(e[0], e[1]): f"{e[2]['lift']:.2f}" for e in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

plt.title("Association Rules Network Graph (Lift)")
plt.show()
```
