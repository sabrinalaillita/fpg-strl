Data Wrangling

```python
#pip install pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# import dataset
df = pd.read_csv("Sales.csv")
print(df.head(5))
```

```python
df.describe(include="all")
```

```python
df.info()
```

```python
# Change 'Tanggal' data type to Datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
df.info()
```

```python
# Deleteing rows with 'status' = 'belum lunas'
df_clean = df.copy()
print(df_clean[df_clean['Status'].str[:1]=='B'].count())
df_clean = df_clean[df_clean['Status'].str[:1]!='B']
```

```python
# Deleteing rows with 'HeaderReturFaktur' > 0
print(df_clean[df_clean['HeaderReturFaktur']>0].count())
df_clean = df_clean[df_clean['HeaderReturFaktur']==0]
```

```python
df_clean.info()
```

Data Cleansing

```python
# Check for missing value
df_clean.isnull().sum()
```

```python
# Check for duplicated data
print("Jumlah duplikasi: ", df_clean.duplicated().sum())
```

```python
# Resolving data inconsistencies

df_clean['Outlet'] = df_clean['Outlet'].str.title()
df_clean['NomorFaktur'] = df_clean['NomorFaktur'].str.upper()
df_clean['Status'] = df_clean['Status'].str.title()
df_clean['SKU_Produk'] = df_clean['SKU_Produk'].str.upper()
df_clean['NamaProduk'] = df_clean['NamaProduk'].str.title()
```

Exploratory Data Analysis (EDA)

```python
# Find the 5 top-selling products
# Group by product and sum the quantities
sales_by_product = df_clean.groupby('NamaProduk')['Qty'].sum()

# Sort the products by total sales and select the top 5
top_5_products = sales_by_product.sort_values(ascending=False).head(5)
print(top_5_products)
```

```python
# Find the worst-selling products
# Group by product and sum the quantities
sales_by_product = df_clean.groupby('NamaProduk')['Qty'].sum()

# Sort the products by total sales and select the top 5
worst_selling_products = sales_by_product[sales_by_product <= 1]
print(worst_selling_products)
```

```python
# Find the top-selling Outlet
# Group by product and sum the quantities
sales_by_outlet = df_clean.groupby('Outlet')['Qty'].sum()

# Sort the outlets by total sales
sorted_sales_by_outlet = sales_by_outlet.sort_values(ascending=False)

# Best-selling outlet
best_selling_outlet = sorted_sales_by_outlet.idxmax()
print("Best-selling Outlet: " + best_selling_outlet)

# Worst-selling outlet
worst_selling_outlet = sorted_sales_by_outlet.idxmin()
print("Worst-selling Outlet: " + worst_selling_outlet)
```

Visualization & Explanatory Analysis

```python
# mengubah frekuensi data -- melihat jumlah order dan total revenue yang diperoleh setiap bulannya
monthly_orders_df = df_clean.resample(rule='ME', on='Tanggal').agg({ #rule = M(Monthly), on = pada kolom
    "NomorFaktur": "nunique",
    "HeaderTotalFaktur": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "NomorFaktur": "order_count",
    "HeaderTotalFaktur": "revenue"
}, inplace=True)
monthly_orders_df.head()

```

```python
# membuat line chart jumlah order per bulan
plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_df["Tanggal"], monthly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Month (2016-2018)", loc="center", fontsize=20)
plt.xticks(rotation=45)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(5000,9000)
plt.show()
```

```python
# How has the company's sales and revenue performed?

weekly_orders_df = df_clean.resample(rule='W', on='Tanggal').agg({
    "NomorFaktur": "nunique",
    "HeaderTotalFaktur": "sum"
})
weekly_orders_df.index = weekly_orders_df.index.strftime('%Y-%m-%d')
weekly_orders_df = weekly_orders_df.reset_index()
weekly_orders_df.rename(columns={
    "NomorFaktur": "order_count",
    "HeaderTotalFaktur": "revenue"
}, inplace=True)
weekly_orders_df.head()
```

```python
# Membuat line chart jumlah order per minggu
plt.figure(figsize=(10, 5))
plt.plot(weekly_orders_df["Tanggal"], weekly_orders_df["order_count"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Week (2016-2018)", loc="center", fontsize=20)
plt.xticks(rotation=45)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)
plt.show()
```

```python
# membuat line chart revenue per bulan
plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_df["Tanggal"], monthly_orders_df["revenue"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Total Revenue per Month (Oct 2023 - Feb 2024)", loc="center", fontsize=18)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.ylim(bottom=2000000000) # Batas bawah 2 miliar
plt.show()
```

Persiapan data untuk Market Basket Analysis

```python
basket = pd.pivot_table(df_clean, index='NomorFaktur', columns='NamaProduk', values='SKU_Produk', aggfunc='nunique', fill_value=0)
basket
```

```python
# Encode Dataframe basket dengan True untuk nilai lebih dari 0 dan False untuk semua nilai 0

def encode(x):
    if x==0:
        return False
    if x>0:
        return True

basket_encode = basket.applymap(encode)
basket_encode
```

```python
# Ambil transaksi yang banyak produk unik lebih dari 1 aja
basket_filter = basket_encode[(basket_encode>0).sum(axis=1)>1]
basket_filter
```

```python
basket_filter.info()
```

Market Basket Analysis menggunakan Frequent Pattern Growth

```python
#pip install mlxtend
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Nilai support 0.01
frequent_itemsets = fpgrowth(basket_filter, min_support=.01, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets['product_cnt'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold=.7).sort_values(['support','confidence'], ascending=[False,False]).reset_index(drop=True)
rules
```
